```python
# Bar graph Brand Vs No. of Laptops (using plotly)
x = laptops_df['Brand'].value_counts().index
y = laptops_df['Brand'].value_counts().values

df = pd.DataFrame({'Brand':x,
                  'No':y })

fig = px.bar(df, 
             x='Brand', 
             y='No',
             color='Brand',
             title='Brand Vs No. of Laptops'
            )
fig.show()
```

```python
# Bar graph Brand Vs Most Expensive Laptop (using plotly)
expensive = laptops_df.groupby('Brand').max('Sellig Price')
x = expensive['Sellig Price'].index
y = expensive['Sellig Price'].values

df = pd.DataFrame({'Brand':x,
                  'Selling Price':y })

fig = px.bar(df, 
             x='Brand', 
             y='Selling Price',
             color='Brand',
             title='Brand Vs Most Expensive Laptop'
            )
fig.show()
```

```python
# Bar graph Brand Vs Least Expensive Laptop (using plotly)
cheapest = laptops_df.groupby('Brand').min('Sellig Price')
x = cheapest['Sellig Price'].index
y = cheapest['Sellig Price'].values

df = pd.DataFrame({'Brand':x,
                  'Selling Price':y })

fig = px.bar(df, 
             x='Brand', 
             y='Selling Price',
             color='Brand',
             title='Brand Vs Least Expensive Laptop'
            )
fig.show()
```

```python
def convert_rating(str1):
    new_str = str1.split(" ")[0]
    return int(new_str)
    
laptops_df['nor'] = laptops_df['No_of_ratings'].apply(convert_rating)
```

```python
laptops_df.head()
```

```python
# Laptop with highest ratings based on above mean no of ratings
laptops_rating = laptops_df[(laptops_df['nor'] >= laptops_df['nor'].mean())]
laptops_rating[(laptops_rating['Ratings'] == laptops_rating['Ratings'].max())]
```

```python
# Laptop with lowest ratings based on above mean no of ratings
laptops_rating = laptops_df[(laptops_df['nor'] >= laptops_df['nor'].mean())]
laptops_rating[(laptops_rating['Ratings'] == laptops_rating['Ratings'].min())]
```
